1. Load the necessary package for plotting using pyplot from matplotlib. 
Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. 
Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]
         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6,7,8]
y = [60,62,61,58,56,57,46,63]
plt.plot(x, y)
plt.xlabel('Days')
plt.ylabel('Car-speed')
plt.title('Car Speed Measurement')
plt.show()  
```

```python
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='+', markerfacecolor='Red', markersize=20)
```

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sine')
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title(' Cosine')
plt.show()
```

6.  Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students by marks range and try to figure out how many Students are average, below-average and Excellent.

Marks = [ 61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]

Histogram showing Below Average, Average and Execellent distribution

40-60: Below Average

60-80: Average

80-100: Excellent

```python
mark = [61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64]
bins = np.arange(40,100)
plt.hist(mark, 30, range=[40, 60], facecolor='blue', align='mid', label='Below Average')
plt.hist(mark, 30, range=[60, 80], facecolor='yellow', align='mid', label='Average')
plt.hist(mark, 30, range=[80, 100], facecolor='red', align='mid', label='Excellent')
plt.xlabel('Number of students')
plt.ylabel('Marks of students')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('/content/titanic.csv')
print(data.head())
```

```python
data.columns.values
```

```python
nam = data['Sex'].unique()
nam
```

```python
dat = data['Sex'].value_counts()
dat
```

```python
col= ['green','yellow']
plt.pie(dat,labels=nam,colors=col,startangle=180, shadow=True, autopct='%1.2f%%')
plt.title('Male/Female Proportion')
plt.show()
```

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
d = {"Age"      : data['Age'],
     "Fare"    : data['Fare'],
     "Sex" : data['Sex']}
df = pd.DataFrame(d)
print(df)
```

```python
colors = np.where(df["Sex"]=='male','Red','black')
df.plot.scatter(x="Fare",y="Age",c=colors)
plt.title('Male/Female Age&Fare Proportion')
plt.show()
```
